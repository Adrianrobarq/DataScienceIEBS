El siguiente conjunto de datos (day.csv) contiene información acerca del uso de un servicio de alquiler de bicicletas.

Calcula la correlación de las variables “temp”, “atemp”, “hum”, y “windspeed” con la variable objetivo “cnt”. ¿Cuál presenta más correlación? Interpreta los resultados obtenidos.



```python
#Primero importamos las librerías necesarias
import pandas as pd
import numpy as np
import scipy as sp

#Importamos librerias para cargar datos en colab
from google.colab import files
import io
```

```python
from os import name
#Cargamos los datos desde local
uploaded = {}
uploaded = files.upload()
```

```python
#Leemos el nombre del archivo, por si se carga varias veces.
names = []
for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
  names.append(fn)
```

```python
df_data = pd.read_csv(names[0])

df_data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Explicación de las variables:


*   temp: Temperatura normalizada registrada, medida en Celsius.
*   atemp: Sensación térmica normalizada, en Celsius.
*   hum: Humedad, en tanto por 1.
*   windspeed: Velocidad del viento, km/h dividido por 67.
*   cnt: Conteo del total de alquileres de bicicletas.



```python
#Calculamos la correlación entre las variables solicitadas y presentamos un mapa
correl = df_data[['temp', 'atemp', 'hum', 'windspeed']].corrwith(df_data['cnt'])

print(correl)

corr_cruz = df_data[['temp', 'atemp', 'hum', 'windspeed']].corr()

sns.heatmap(corr_cruz, annot=True, cmap='coolwarm')
```
