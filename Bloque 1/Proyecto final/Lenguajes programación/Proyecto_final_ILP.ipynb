## 5. Asistencia de público

La columna `Attendance` indica la asistencia de público en cada mundial. Pandas ha cargado esta columna como tipo string, ya que los valores indicados contienen puntos como separador de millares. En primer lugar vamos a transformar esta columna para que no contenga el caracter `.`. Posteriormente transformaremos esta columna a formato numérico usando la función [pd.to_numeric](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.to_numeric.html).

**Nota**: la función `pd.read_csv` también permite hacer este proceso de una forma mucho más transparente, usando el argumento `thousands`. Ver la documentación [aquí](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html).

````python
df_copas["Attendance"] = pd.to_numeric(df_copas["Attendance"].str.replace(".", ""))
````

````python
df_copas.dtypes
````

Teniendo en cuenta todos los datos de `df_copas`, ¿Cuál fué la mayor asistencia de un mundial? ¿Cuál es la asistencia media?

````python
# Completar

asistencia_max = df_copas["Attendance"].max()
asistencia_media = df_copas["Attendance"].mean()


country_max = df_copas.set_index('Attendance').at[asistencia_max,'Country']
year_max = df_copas.set_index('Attendance').at[asistencia_max,'Year']

print('La asistencia media ha sido de {} personas entre los años {} y {}'.format(
    asistencia_media, primero, ultimo))
print('La asistencia máxima ha sido de {} personas durante el mundial de {} en {}'.format(
    asistencia_max, country_max, year_max))
````

## 11. Número de victorias por país

Calcular para cada país el número de partidos ganados. Para ello, usar la función [dropna](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html) para eliminar las filas correspondientes a partidos donde la columna `Ganador` es un valor perdido (`na`). Ordenar los países por número de victorias de forma descendente.

````python
# Completar
#Extraemos el dato de todos los paises que han jugado partidos
paises = partidos_jugados.index

#Creamos el dataframe de victorias por pais
victorias_por_pais = df_partidos.dropna().groupby('Ganador')['Year'].count().sort_values(ascending = False)

#Añadimos los paises con 0 victorias
for pais in paises:
  if pais not in victorias_por_pais.index:
    victorias_por_pais = pd.concat([victorias_por_pais,pd.Series(data = [0],
                                                                 index = [pais])])

#Renombramos, importante para el merge posterior
victorias_por_pais = victorias_por_pais.rename('Partidos Ganados')

victorias_por_pais
````

En primera aproximación, si se inspeccionaban los datos tras realizar el conteo y la agreupación, veíamos que obteníamos un dataframe con 64 filas, mientras que el de partidos jugados tenía un total de 83. Esto se debe a que al agrupar por ganador, los paises que no habían ganado una sola vez no aparecían en el conteo. De esta forma añadimos, a partir de los datos extraidos previamente en el dataframe `partidos_jugados` los paises restantes y les asignamos el valor de 0 victorias. Esto ayudará en la siguiente fase, ya que se puede comprobar que la función `pd.merge` unicamente cruzará los datos compartidos, por lo que es importante que los dos dataframes contengan la misma cantidad de filas para no perder información.

## Opcional 1

Calcular para cada país el ratio de victorias, calculado como el número de partidos ganados dividido por el número de partidos jugados. Ordenar los países por ratio de forma descendente.

**Pista:** Si obtenemos dos dataframes, uno con las victorias por cada país y otro con el número total de partidos jugados, podemos utilizar la función [pd.merge](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.merge.html) para combinar ambos en un solo DataFrame, y despues crear una nueva columna que sea el ratio.

````python
# Completar

#Creamos el dataframe a partir de la fusión
#Los paises que nunca hayan ganado un partido no aparecerán
victorias_info = pd.merge(victorias_por_pais, partidos_jugados, right_on = 'Country'
, left_index= True).rename(columns = {'Year':'Partidos Ganados'})

#Creamos la columna Ratio
victorias_info['Ratio'] = victorias_info['Partidos Ganados']/victorias_info['Partidos Jugados']

victorias_info.sort_values('Ratio', ascending = False)
````

Para operar con las Series, el tipo de tabla que surge al realizar el count, es importante nombrarlas de forma que la función `pd.merge` pueda operar y alinear las columnas adecuadamente.
